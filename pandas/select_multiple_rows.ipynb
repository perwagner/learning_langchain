```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry", index_col="country")
drinks.head()
```

```python
drinks.loc["Denmark", ["wine_servings", "continent"]]
```

```python
drinks.loc[["Denmark", "Finland"], ["wine_servings", "continent"]]
```

```python
drinks.loc["Denmark", drinks.columns[1: 4]]
```

```python
drinks.iloc[drinks.index.get_loc("Denmark"), :]
```

```python
drinks.iloc[5, drinks.columns.get_loc("continent")]
```
